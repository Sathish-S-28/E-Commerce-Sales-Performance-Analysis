```python

products["Category_name"] = products["Category_name"].str.lower().str.strip()

def final_clean(x):

    
    if "food" in x or "drink" in x:
        return "food"

    
    elif any(k in x for k in ["fashion", "clothing", "shoes", "bags"]):
        return "fashion"

    
    elif any(k in x for k in ["computer", "electronics", "phone", "audio", "tablet", "telephony"]):
        return "electronics"

    
    elif any(k in x for k in ["home", "appliance", "kitchen", "comfort", "housewares"]):
        return "home"

    
    elif any(k in x for k in ["furniture", "bed", "sofa", "table"]):
        return "furniture"

    
    elif any(k in x for k in ["book", "cd", "dvd", "music"]):
        return "books_media"

    
    elif any(k in x for k in ["beauty", "health", "perfume"]):
        return "beauty_health"

    
    elif any(k in x for k in ["toy", "baby", "diaper"]):
        return "toys_baby"

    
    elif "sport" in x:
        return "sports"

    
    elif any(k in x for k in ["auto", "car"]):
        return "automotive"

    
    elif any(k in x for k in ["construction", "tools", "industry", "security"]):
        return "industrial"

    
    elif "garden" in x:
        return "garden"

    
    elif any(k in x for k in ["office", "stationery"]):
        return "office"

    
    elif "pet" in x:
        return "pet"

    
    elif any(k in x for k in ["cool", "gift"]):
        return "gifts"

    
    elif any(k in x for k in ["art", "craft"]):
        return "art_craft"

    
    elif "luggage" in x:
        return "travel"

    
    elif any(k in x for k in ["party", "christmas"]):
        return "events"

    
    elif any(k in x for k in ["photo", "cine"]):
        return "photography"

    
    elif "market" in x:
        return "marketplace"

    
    elif "air" in x:
        return "home"

    
    else:
        return x



products["Category_final"] = products["Category_name"].apply(final_clean)


print("\nFinal Category Count:\n")
print(products["Category_final"].value_counts())
```

```python
'''
                                              T-Test

Do loyal customers spend more revenue than new customers?

H0: Mean revenue is the same for loyal and new customers
H1: Mean revenue is different
'''
```

```python
from scipy import stats

data = order_items_cln.merge(orders_cln[['order_id','customer_id']], on='order_id') \
                      .merge(customers_cln[['customer_id','customer_segment']], on='customer_id')

loyal = data[data['customer_segment'] == 'Loyal']['net_revenue'].dropna()
new = data[data['customer_segment'] == 'New']['net_revenue'].dropna()

print("Loyal Customer Mean Revenue:", loyal.mean())
print("New Customer Mean Revenue:", new.mean())

t_stat, p_value = stats.ttest_ind(loyal, new)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Decision: Reject H0")
    print("Conclusion: Loyal and new customers have different revenue")
else:
    print("Decision: Fail to Reject H0")
    print("Conclusion: Loyal and new customers spend the same revenue")
```

```python
'''
                                                   Z Test

Is the average review score different from 4.1?

H0: Mean review score = 4.0
H1: Mean review score ≠ 4.0
'''
```

```python
import numpy as np
from scipy.stats import norm

review = reviews_cln['review_score'].dropna()

sample_mean = review.mean()
sample_std = review.std()
n = len(review)

print("Average Review Score:", sample_mean)

mu = 4.0

z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Decision: Reject H0")
    print("Conclusion: The average review score is significantly different from 4.0")
else:
    print("Decision: Fail to Reject H0")
    print("Conclusion: The average review score is approximately 4.0")
```

```python
'''
                                              Chi-Square Test

Is customer segment related to order status?

H0: Customer segment and order status are independent
H1: Customer segment and order status are related
'''
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = orders_cln.merge(customers_cln[['customer_id','customer_segment']], on='customer_id')

table = pd.crosstab(data['customer_segment'], data['order_status'])

chi2, p_value, dof, expected = chi2_contingency(table)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Decision: Reject H0")
    print("Conclusion: Customer segment and order status are related")
else:
    print("Decision: Fail to Reject H0")
    print("Conclusion: Customer segment and order status are independent")
```

```python
'''
                                                    ANOVA

Do different age groups generate different revenue levels?

H0: Mean revenue is the same across all age groups
H1: At least one age group has a different mean revenue
'''
```

```python
from scipy import stats

data = order_items_cln.merge(orders_cln[['order_id','customer_id']], on='order_id') \
                      .merge(customers_cln[['customer_id','age_group']], on='customer_id')

groups = [group['net_revenue'].dropna() for name, group in data.groupby('age_group')]

f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Decision: Reject H0")
    print("Conclusion: At least one age group generates different revenue")
else:
    print("Decision: Fail to Reject H0")
    print("Conclusion: All age groups generate similar revenue")
```

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score



# merge datasets

data = reviews_cln.merge(order_items_cln[['order_id','shipping_cost']], on='order_id')



# target variable

data["good_review"] = data["review_score"].apply(lambda x: 1 if x >= 4 else 0)



# feature & target

X = data[['shipping_cost']]
y = data['good_review']



# handle missing values

data = data.dropna(subset=['shipping_cost','good_review'])



# scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



# train test split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)



# find best k

error_rate = []
k_values = range(1,15)
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred_k = knn.predict(X_test)
    error_rate.append(np.mean(pred_k != y_test))



# error rate vs k

plt.figure(figsize=(6,4))
plt.plot(k_values, error_rate, marker="o")
plt.title("Error Rate vs K")
plt.xlabel("K")
plt.ylabel("Error Rate")
plt.show()



# best k

best_k = k_values[np.argmin(error_rate)]
print("Best K:", best_k)



# final model

model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train, y_train)



# prediction

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]



# evaluation

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Precision:", precision_score(y_test,y_pred))
print("Recall:", recall_score(y_test,y_pred))
print("F1 Score:", f1_score(y_test,y_pred))
print("ROC AUC:", roc_auc_score(y_test,y_prob))

print("\nClassification Report\n")
print(classification_report(y_test,y_pred))



# confusion matrix

cm = confusion_matrix(y_test,y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()



# roc auc
fpr, tpr, _ = roc_curve(y_test,y_prob)

plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label="ROC Curve")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score



# select features

X = order_items_cln[['unit_price','shipping_cost']]



# handle missing values

X = X.dropna()



# scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



# elbow method

wss = []

for k in range(1,15):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wss.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1,15), wss, marker='o')
plt.title("Elbow Method")
plt.xlabel("K")
plt.ylabel("WSS")
plt.show()



# final model

k = 3
kmeans = KMeans(n_clusters=k, random_state=42)

clusters = kmeans.fit_predict(X_scaled)

# add cluster to dataset
order_items_cln = order_items_cln.loc[X.index]
order_items_cln["cluster"] = clusters



# visualization

sample_df = order_items_cln.sample(20000, replace=True)

plt.figure(figsize=(6,4))
sns.scatterplot(
    x=sample_df["unit_price"],
    y=sample_df["shipping_cost"],
    hue=sample_df["cluster"]
)
plt.title("K-Means Clusters")
plt.show()



# metrics

sample_X = X_scaled[np.random.choice(len(X_scaled),20000,replace=True)]
sample_clusters = kmeans.predict(sample_X)

print("Silhouette Score:", silhouette_score(sample_X, sample_clusters))
print("Calinski-Harabasz Index:", calinski_harabasz_score(sample_X, sample_clusters))
print("Davies-Bouldin Index:", davies_bouldin_score(sample_X, sample_clusters))
```
